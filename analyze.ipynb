```python
%matplotlib inline
from lightjob.cli import load_db
from IPython.display import Image, display
from skimage.io import imsave
import os
```

```python
db = load_db()
```

```python
jobs = db.jobs_with(state='success')
```

```python
def key(j):
    return db.get_value(j, 'stats.out_of_the_box_classification.letterness.diversity_max_letters', if_not_found=None)

jobs = [j for j in jobs if j['stats']]
jobs = sorted(jobs, key=key)
jobs = jobs[::-1]
print(map(key, jobs))
```

```python
for j in jobs[0:10]:
    folder = 'jobs/{}/samples/'.format(j['summary'])
    sample = max(os.listdir(folder), key=lambda f:int(f[0:5]))
    display(Image(folder + '/' + sample))
```
